```python
import csv

csv_file = 'Final_database_to_upload.csv'

with open(csv_file, newline='') as file:
    reader = csv.DictReader(file)
    
    for i, row in enumerate(reader):
        if i >= 100:
            break
        
        contest_index = row.get('Contest', '')
        contest_index = contest_index[:-2]
        probname=row.get('Problem','')
        index = row.get('Index', '')
        tag = row.get('Tags', '')
        solved_count = row.get('solvedCount', '')
        
        output = f'<tr>\n'
        output += f'    <th scope="row">{i + 1}</th>\n'
        output += f'    <td><a href="https://codeforces.com/problemset/problem/{contest_index}/{index}" target="_blank" class="text-primary">{probname}</a></td>\n'
        output += f'    <td>{tag}</td>\n'
        output += f'    <td>{solved_count}</td>\n'
        output += f'    <td><span class="badge bg-success">Approved</span></td>\n'
        output += f'</tr>'
        
        print(output)
```
