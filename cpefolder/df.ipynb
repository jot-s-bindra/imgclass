```python
import requests
import csv

codeforces_api_url = "https://codeforces.com/api/user.ratedList"

min_rating = 1200

def get_users_with_high_ratings(min_rating):
    params = {
        "activeOnly": "true",
    }

    # Send a GET request to the Codeforces API
    response = requests.get(codeforces_api_url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        if data["status"] == "OK":
            # Filter users with a maximum rating greater than min_rating
            high_rating_users = [user for user in data["result"] if user["maxRating"] > min_rating]
            return high_rating_users
        else:
            print("Codeforces API request failed:", data["comment"])
    else:
        print("Failed to connect to the Codeforces API. Check your internet connection or try again later.")

# Define a function to write the data to a CSV file
def write_to_csv(users, filename):
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Handle", "Max Rating"])
        for user in users:
            writer.writerow([user['handle'], user['maxRating']])

if __name__ == "__main__":
    high_rating_users = get_users_with_high_ratings(min_rating)
    if high_rating_users:
        print("Users with a maximum rating greater than", min_rating)
        for user in high_rating_users:
            print(f"{user['handle']} (Max Rating: {user['maxRating']})")

        # Save the data to a CSV file
        csv_filename = "high_rating_users.csv"
        write_to_csv(high_rating_users, csv_filename)
        print(f"Data has been saved to {csv_filename}")
```
