```python
_dict = {
    'A1': 'A',
    'A2': 'A',
    'B1': 'B',
    'B2': 'B',
    'C1': 'C',
    'C2': 'C',
    'D1': 'D',
    'D2': 'D',
    'D3': 'D',
    'E1': 'E',
    'E2': 'E',
    'F1': 'F',
    'F2': 'F',
    'G1': 'G',
    'G2': 'G',
    'H2': 'H',
    'F3': 'F'
}

df['Index'] = df['Index'].replace(_dict)
```

```python
df['Index'].unique()
```

```python
df['Index'] = df['Index'].astype('category')
df['Index_New'] = df['Index'].cat.codes
```

```python
count = 0
for index,row in df.iterrows():
    if 'implementation' in row['Tags']:
        count+=1
print(count)
```

```python
_List = ['greedy',
         'math',
         'brute force',
         'constructive algorithms',
         'data structure',
         'dp',
         'string',
         'sorting',
         'binary search',
         'dfs',
         'two pointers',
         'graph',
         'special',
         'combinatorics',
         'number theory',
         'tree']
```

```python
for item in _List:
    df[item] = 0
    for index,row in df.iterrows():
        if item in row['Tags']:
            df.loc[index, item] = 1

df.head()
```

```python
df = df.drop(labels='Tags', axis=1, inplace=False)
df = df.drop(labels='Index', axis=1, inplace=False)
df = df.drop(labels='Problem', axis=1, inplace=False)
df = df.drop(labels='Contest', axis=1, inplace=False)
```
